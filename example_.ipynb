# Example for using the library seriation_mathopt

Load library and a matrix. In this case the GoT Season 1 character interaction matrix:

```python
import numpy as np
from seriation_mathopt import *

A = np.loadtxt("experiments/real_instances/got1_50.txt", delimiter=",")
np.fill_diagonal(A, 1)
names=['ALLISER_THORNE', 'ARYA', 'BARRISTAN', 'BENJEN', 'BRAN', 'BRONN', 'CATELYN', 'CERSEI', 'DAENERYS', 'DOREAH', 'DROGO', 'GREATJON_UMBER', 'GRENN', 'HOUND', 'ILLYRIO', 'IRRI', 'JAIME', 'JEOR', 'JOFFREY', 'JON', 'JORAH', 'JORY_CASSEL', 'LITTLEFINGER', 'LORAS', 'LYSA', 'MAESTER_AEMON', 'MAESTER_LUWIN', 'MIRRI_MAZ_DUUR', 'NED', 'OSHA',
'PYCELLE', 'PYP', 'QOTHO', 'RAKHARO', 'RENLY', 'ROBB', 'ROBERT', 'ROBIN', 'RODRIK', 'ROS', 'SAM', 'SANSA', 'SEPTA_MORDANE', 'SHAE', 'THEON', 'TYRION', 'TYWIN', 'VARYS', 'VISERYS', 'YOREN']
```

Run the code with the desired parameters: method, symmetric_order, neighbors, ...

```python
seriation_example1 = MatrixSeriation(A, method="tsp", coordinated_ordering=False, eps_neigh="VonNeumann", output=0) #VonNeuman neighborhood
seriation_example1 .solve()
seriation_example1.plot_matrices(names=names)
```

```python
seriation_example2 = MatrixSeriation(A, method="tsp", coordinated_ordering=False, eps_neigh="ME") #MEasure of efficiency
seriation_example2 .solve()
seriation_example2.plot_matrices(names=names)
```
