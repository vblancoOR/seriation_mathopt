```python

#The parameters for the problem are set:

seriation_example = MatrixSeriation(A, method="tsp", symmetric_ordering=True, eps_neigh=1, output=1)

#The solver is called:

example = seriation_example .solve()

#The solution can be extracted

print(seriation_example.solution_info)

#The transformed matrix can be extracted:

print(seriation_example.transformed_matrix)
```
