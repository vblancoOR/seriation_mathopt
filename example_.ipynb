# Example for using the library seriation_mathopt

Load library and a matrix. In this case the GoT Season 1 character interaction matrix:

```python
import numpy as np
from seriation_mathopt import *

matrix_got1 = np.loadtxt("datasets/got1_50.txt", delimiter=",")
np.fill_diagonal(matrix_got1, 1)
names_got1 = np.loadtxt("datasets/got1_50_names.txt", dtype=str)
```

Run the code with the desired parameters: method, symmetric_order, neighbors, ...

```python
seriation_got1 = MatrixSeriation(matrix_got1, method="HPM", coordinated_ordering=True, eps_neigh="ME") #MEasure of efficiency
seriation_got1 .solve()
seriation_got1.plot_matrices(names=names_got1, save=True)
```
